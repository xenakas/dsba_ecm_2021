```r
x1 <- rnorm(200, 30, 15)
x2 <- rnorm(200, 30, 15)
e <- rnorm(200, 0, 10)
y <- x1+e
```

```r
plot(x1,y)
plot(x2,y)
```

```r
summary(lm(y~x1))
summary(lm(y~x2))
```
